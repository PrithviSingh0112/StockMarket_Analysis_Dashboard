```python
import pandas as pd
Apple = pd.read_csv('stocks/AAPL.csv')
facebook = pd.read_csv('stocks/FB.csv')
google = pd.read_csv('stocks/GOOGL.csv')
nvidia = pd.read_csv('stocks/NVDA.csv')
tesla = pd.read_csv('stocks/TSLA.csv')
twitter = pd.read_csv('stocks/TWTR.csv')
adobe = pd.read_csv('stocks/ADBE.csv')
AMD = pd.read_csv('stocks/AMD.csv')
netflix = pd.read_csv('stocks/NFLX.csv')
SBI = pd.read_csv('stocks/SBI.csv')
```

```python
Apple.head()
```

```python
dfs = [Apple,facebook,google,nvidia,tesla,twitter,adobe,AMD,netflix,SBI]
```

```python
for df in dfs:
    df['MA10'] = df.Close.rolling(10).mean()
    df['MA100'] = df.Close.rolling(100).mean()
```

```python
Apple.head(10)
```

```python
for df in dfs:
    df['Previous day close price'] = df.Close.shift(1)
    df['Change in Price'] = df['Close'] - df['Previous day close price']
    df['percentage change in price'] = df.Close.pct_change()
```
